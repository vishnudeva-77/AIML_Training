```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed()
```

```python
employee_ids = [f'EID-{i+1}' for i in range(30)]
```

```python
days = pd.date_range(start='2023-01-01', periods=30, freq='D')
working_hours = np.random.randint(4, 11, size=(30, 30))
```

```python
data = pd.DataFrame(working_hours, index=days, columns=employee_ids)
```

```python
print(data.head())
```

```python
print(data.describe())
```

```python
average_hours = data.mean(axis=0)
print("\nAverage Working Hours per Employee:\n", average_hours)
```

```python
anomalies = data[data > 8].count()
print("\nAnomalies (working hours > 8):\n", anomalies)
```

```python
total_hours = data.sum()
plt.figure(figsize=(12, 6))
plt.bar(total_hours.index, total_hours.values, color='skyblue')
plt.title('Total Working Hours per Employee')
plt.xlabel('Employee IDs')
plt.ylabel('Total Working Hours')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
daily_average_hours = data.mean(axis=1)
plt.figure(figsize=(12, 6))
plt.plot(daily_average_hours.index, daily_average_hours.values, marker='o', linestyle='-')
plt.title('Daily Average Working Hours')
plt.xlabel('Date')
plt.ylabel('Average Working Hours')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
