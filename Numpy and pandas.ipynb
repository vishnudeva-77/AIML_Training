```python
#Analysis of Electricity Usage
```

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
apartments=[f"Apt_{i})" for i in range (1,21)]
```

```python
dates=pd.date_range(start="2025-07-01",periods=31,freq="D")
```

```python
dates
```

```python
data={
    "Date":np.tile(dates,len(apartments)),
    "Apartments":np.repeat(apartments,len(dates)),
    "Electricity_usage":np.random.normal(loc=20,scale=5,size=len(dates)*len(apartments))
}
data
```

```python
df=pd.DataFrame(data)
df["Electricity_usage"]=df["Electricity_usage"].round(2)
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
print(df.dtypes)
```

```python
usage_per_apartments = df.groupby("Apartments")["Electricity_usage"].sum().sort_values(ascending=False)
usage_per_apartments
```

```python
print(df.columns)
```

```python
daily_avg = df.groupby("Date")["Electricity_usage"].mean()
print(daily_avg.head())
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=usage_per_apartments.index,y=usage_per_apartments.values)
plt.xticks(rotation=90)
plt.title("Total Usage per Apartments ")
plt.xlabel("Apartments")
plt.ylabel("Total KWH")
plt.tight_layout()
plt.show()
```

```python
#Time Series Plot of Daily Average
plt.figure(figsize=(10,6))
sns.barplot(x=daily_avg.index,y=daily_avg)
plt.xticks(rotation=90)
plt.title("Daily average Usage per Apartments ")
plt.xlabel("Date")
plt.ylabel("Average KWH")
plt.tight_layout()
plt.show()
```

```python
threshold = df["Electricity_usage"].mean() + 2 * df["Electricity_usage"].std()
df["High_usage_flag"] = df["Electricity_usage"] > threshold
print(df[df["High_usage_flag"] == True].head())
```
